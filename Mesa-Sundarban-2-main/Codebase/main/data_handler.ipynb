```python
# Now we combine these all to a general function to use later
# property is a string 

def process_data(property,start_year,end_year):
    property = property.lower()
    df = pd.read_csv('./dataset/public/' + property + '.csv')
    stations = ['Khulna','Satkhira','Sandwip','Patuakhali'] # Related to the Sundarbans
    df = df[df['Year'] >= start_year]
    df = df[df['Year'] <= end_year]
    df = df[df['Station'].isin(stations)]
    drop_columns = ['29','30','31'] # Getting rid of columns with null values since they will not be much significant
    df.drop(columns=drop_columns,inplace=True)
    days = []
    for i in range(28):
        days.append(str(i+1))
    for day in days:
        df[day] = pd.to_numeric(df[day], errors='coerce').astype(float)
    df[property] = df[days].mean(axis=1)
    df.drop(columns=days,inplace=True)  # Proceeding with monthly average data
    window_size = 5
    rolling_avg = df[property].rolling(window_size, min_periods=1).mean()
    df[property] = df[property].fillna(rolling_avg)
    df['Year'] = df['Year'].astype(int)
    df['Month'] = df['Month'].astype(int)
    df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')
    # Set the 'Date' column as the index
    df.set_index('Date', inplace=True)
    df = df.sort_index()
    df.drop('Station',axis=1,inplace=True)
    df = df.groupby([df.index.year, df.index.month]).mean()
    newdf = df.reset_index(drop=True)
    newdf['Year'] = newdf['Year'].astype('int')
    newdf['Month'] = newdf['Month'].astype('int')
    return newdf
```

```python
def get_climate_data(properties,start_year,end_year):
    df = process_data(properties[0],start_year,end_year)
    for property in properties[1:]:
        #print(property)
        ret = process_data(property,start_year,end_year)
        df[property] = ret[property]
    for property in properties:
        #print(property.capitalize())
        df.rename(columns={property : property.capitalize()},inplace=True)
    return df
```

```python
properties = ['temparature','humidity','rainfall','sunshine'] # Small Letter is must, must match with file name
start_year = 1990
end_year = 2010
df = get_climate_data(properties,start_year,end_year)

df
```

```python
# Sunshine data tends to be missing for some years
# For 1990 to 2010, all data exist
df.isna().sum()
```

```python
def generate_climate_data_csv(start_year,end_year):
    properties = ['temparature','humidity','rainfall','sunshine'] # Small Letter is must, must match with file name
    df = get_climate_data(properties,start_year,end_year)
    filename = 'climate_'+str(start_year)+'-'+str(end_year)
    df.to_csv('./dataset/generated/'+filename+'.csv',index=False)
    print('Dataset Successfully Generated')
```

```python
generate_climate_data_csv(1990,2010)
```
